# # basics of grouping

```python
#unique method
df["drive-wheels"].unique()
```

```python
#group by

df_group_one = df[["drive-wheels","body-style","price"]]
df_group_one  =df_group_one.groupby(["drive-wheels"], as_index = True).mean()
df_group_one
```

```python
#grouping results with multiple variable
df_multi_group = df[["drive-wheels","body-style","price"]]
df_multi_group =df_multi_group.groupby(["drive-wheels","body-style"], as_index=False).mean()
df_multi_group
```

```python
#pivoting data for clear visibility
grouped_pivot = df_multi_group.pivot(index= "drive-wheels", columns = "body-style") 
grouped_pivot
```

```python
#to fill misisng values with 0
grouped_pivot.fillna(0)
```

```python
body_style_group = df[["body-style","price"]]
body_style_group = body_style_group.groupby(["body-style"], as_index=False).mean()
body_style_group
```

```python
#using heatmap to visualise relationship between body style and price

plt.pcolor(grouped_pivot, cmap='RdBu')
plt.colorbar()
plt.show()
```
