# Data Wrangling


```python
df.head(5)
```

```python
#to find the missing value in a dataframe

missing_data = df.isnull()
missing_data.head()
```

```python
#count missing value in each column
for column in  missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

```python
#replace missing vales with mean
avg_loss= df["normalized-losses"].astype(float).mean(axis=0) #mean calculated
df["normalized-losses"].replace(np.nan, avg_loss, inplace= True) #replaced
df["normalized-losses"].isnull().value_counts() #to check if replaced correctly
```

```python
#similarly for other columns
avg_bore = df["bore"].astype(float).mean(axis=0)
df["bore"].replace(np.nan, avg_bore, inplace=True)

#replacing stroke
avg_stroke = df["stroke"].astype(float).mean(axis=0)
df["stroke"].replace(np.nan, avg_stroke, inplace =True)
df["stroke"].isnull().value_counts()
```

```python
avg_horsepower = df["horsepower"].astype(float).mean(axis=0)
avg_peakrpm = df["peak-rpm"].astype(float).mean(axis=0)
df["horsepower"].replace(np.nan, avg_horsepower, inplace=True)
df["peak-rpm"].replace(np.nan, avg_peakrpm, inplace=True)
print(df["peak-rpm"].isnull().value_counts())
```

```python
#to replace the values with max value

max_numdoor = df["num-of-doors"].value_counts().idxmax()
df["num-of-doors"].replace(np.nan, max_numdoor, inplace=True)
df["num-of-doors"].isnull().value_counts()
```

```python
#drop all the rows that does not have price
df.dropna(subset=["price"], axis=0, inplace=True)

#resetting index post dropping the columns
df.reset_index(drop=True, inplace=True)
```

# #correcting data formats

```python
#checking datatypes of columns
df.dtypes
```

```python
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

```python
df.dtypes
```

# Data Standardization

```python
#data is to be standardised to keep the value in standard formats
#to convert mpg(miles per gallon) to L/100Km

df["city-L100Km"] = 235/df['city-mpg']
df["highway-L100Km"]= 235/df["highway-mpg"]
```

# Data Normalization

```python
#data is normalised for scaling. To bring data in comparisonal values
#approach is original/max

df["length"] = df["length"]/df["length"].max()
df["width"] = df["width"]/df["width"].max()
df["height"] = df["height"]/df["height"].max()

#all the height, width and length are normalizes and now in between(0,1)
```

# Binning

```python
#binning is used to convert continous numerical values to grouped data[in ranges]
#data binning using cut method in pandas

#first convert data into correct format

df["horsepower"]=df["horsepower"].astype(int, inplace= True)
```

```python
#plotting histogram of horsepower to see the distribution

import matplotlib.pyplot as plt
plt.hist(df["horsepower"])
#setting labels and title
plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("horsepower")
#plt.show()
```

```python
#creating bins
#creating 3 bins of equal values

bins  = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

```python
#set 3 group names
group= ["low", "medium", "High"]

#to add binned data in dataframe

df["binned_horsepower"] = pd.cut(df["horsepower"], bins, labels= group, include_lowest= True)
df[["horsepower", "binned_horsepower"]]
```

```python
#seeing value counts
df["binned_horsepower"].value_counts()
```

```python
#plotting them in bar grapgh
plt.bar(group, df["binned_horsepower"].value_counts())
plt.xlabel("horsepower")
plt.ylabel("counts")
plt.title("horsepower bins")
```

```python
#bins visualization

plt.hist(df["horsepower"],bins=3)
plt.xlabel("horsepower")
plt.ylabel("counts")
plt.title("horsepower")
```

```python
#indicator variables: also dummy variables which are used to convert categorical data to numerical data(used to lable categories)
#as models understand data only. and we can user categorical variable for regression analysis
df.columns

```

```python
#get_dummies method is used to convert categorical to numberical values

dummy_variable_1 = pd.get_dummies(df["fuel-type"])
dummy_variable_1.head()
```

```python
dummy_variable_1.rename(columns={"diesel":"fuel-type-diesel" , "gas":"fuel-type-gas"}, inplace=True)
dummy_variable_1.head()
```

```python
#to merge dummy to orignal dataframe
df= pd.concat([df,dummy_variable_1], axis= 1)
df.head()
```

```python
#to drop original fuel type column

df.drop("fuel-type", axis = 1, inplace= True)
df.head()
```

```python
dummy_variable_2 = pd.get_dummies(df["aspiration"])
dummy_variable_2.head()
```

```python
#renaming 
dummy_variable_2.rename(columns={"turbo":"aspiration-turbo","std":"aspiration-std" }, inplace = True)
dummy_variable_2.head()
```

```python
#merging into original database
df = pd.concat([df, dummy_variable_2], axis = 1,)
df.head()
```

```python
#dropping original aspiration column
df.drop("aspiration", axis=1, inplace=True)
df.head()
```

```python
#saving into new csv file

df.to_csv("clean_df.csv")
```
