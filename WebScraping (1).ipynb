```python
#to fetch information about more products

#URL whose products needs to be  fetched

url = "https://www.flipkart.com/search?q=sandals+for+women&sid=osp%2Ciko%2C9d5&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_13_na_na_ps&otracker1=AS_QueryStore_OrganicAutoSuggest_1_13_na_na_ps&as-pos=1&as-type=RECENT&suggestionId=sandals+for+women%7CWomen%27s+Flats&requestId=d7e8505c-3d4f-49bd-a5aa-44edb83d0c77&as-backfill=on&sort=popularity"
driver.get(url)
```

```python
sandals = []  #to store name of sandals
sandal_price = []   #to store price of sandals
```

```python
content = driver.page_source
soup = BeautifulSoup(content)
names = soup.findAll("div", attrs = { "class": "_2WkVRV"})
print(len(names))
```

```python
price = soup.findAll("div", attrs = {"class": "_30jeq3"})
print(len(price))
price = price[:-5]
print(len(price))
```

```python
for name in names:
    sandals.append(name.text)
```

```python
for prices in price:
    sandal_price.append(prices.text)
```

```python
df1 = pd.DataFrame({"sandal_name": sandals, "sandal_price": sandal_price})
df1
```

```python
#saving the file in csv format
df1.to_csv("sandals.csv", index= False, encoding= 'utf-8-sig')
```
