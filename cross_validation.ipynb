```python
x= iri.data
y= iri.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size =0.22, random_state= 43)
```

```python
mnb = MultinomialNB()
mnb.fit(x_train, y_train)
pred = mnb.predict(x_test)
print(accuracy_score(pred, y_test))
print(classification_report(pred, y_test), confusion_matrix(pred, y_test))
```

```python
svc= SVC()
svc.fit(x_train, y_train)
pred = svc.predict(x_test)
print(accuracy_score(pred,y_test))
print(classification_report(pred, y_test), confusion_matrix(pred, y_test))
```

```python
kn = KNeighborsClassifier()
kn.fit(x_train, y_train)
pred= kn.predict(x_test)
print(accuracy_score(pred, y_test), confusion_matrix(pred, y_test), classification_report(pred, y_test))
```

```python
dtc= DecisionTreeClassifier()
dtc.fit(x_train, y_train)
pred = dtc.predict(x_test)
print(accuracy_score(pred, y_test), confusion_matrix(pred, y_test), classification_report(pred, y_test))
```
