```python
#drop LSTAT
bos.drop('LSTAT', axis = 1, inplace= True )
```

```python
bos.head()
```

```python
#removing outliers
new_bos  = bos[(z<3).all(axis=1)]
new_bos.head()
```

```python
new_bos.shape
```

```python
bos= new_bos
bos.shape
```

```python
# to take all independent variables
x = bos.iloc[:,0:-1]
#dependent variable
y = bos.iloc[:, -1]
```

```python
x.head()
```

```python
y.head()
```

```python
#train and test splitting data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size =0.33,random_state= 42 )
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
lm= LinearRegression()
lm.fit(x_train, y_train)
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
lm.score(x_train, y_train)
```

```python
bos.columns
```

```python
#value prediction
pred = lm.predict(x_test)
pred
```

```python
print("error:")
print("mean_absolute_error:", mean_absolute_error(y_test, pred))
print("mean_squared_error:", mean_squared_error(y_test, pred))
print("root mean square error", np.sqrt(mean_squared_error(y_test, pred)))
```

```python
from sklearn.metrics import r2_score
print("r2 score", r2_score(y_test, pred))
```
